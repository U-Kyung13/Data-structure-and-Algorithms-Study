채점 결과


정확성: 93.8


합계: 93.8 / 100.0

```python
def solution(citations):
    for h in range(len(citations)+1):
        up = sum([x >= h for x in citations])
        down = sum([x < h for x in citations])
        if up < h or down > h:
            break
    return h - 1
```

```python
solution([3, 0, 6, 1, 5])
```

```python
def solution(citations):
    citations.sort()
    answer = 0
    
    while True:
        if len([x for x in citations if x >= answer]) < answer:
            break
        answer += 1
    return answer-1
```

```python
solution([3, 0, 6, 1, 5])
```
